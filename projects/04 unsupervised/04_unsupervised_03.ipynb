```python
import requests
import numpy as np
from PIL import Image
from io import BytesIO
from IPython.display import display
```

```python
url = 'https://media.githubusercontent.com/media/mariolpantunes/ml101/main/figs/01.jpg'
response = requests.get(url)
img01 = Image.open(BytesIO(response.content))

print(img01.format)
print(img01.size)
print(img01.mode)

display(img01)
```

```python
url = 'https://media.githubusercontent.com/media/mariolpantunes/ml101/main/figs/02.jpg'
response = requests.get(url)
img02 = Image.open(BytesIO(response.content))

print(img02.format)
print(img02.size)
print(img02.mode)

display(img02)
```

```python
url = 'https://media.githubusercontent.com/media/mariolpantunes/ml101/main/figs/03.jpg'
response = requests.get(url)
img03 = Image.open(BytesIO(response.content))

print(img03.format)
print(img03.size)
print(img03.mode)

display(img03)
```

```python
data01 = np.asarray(img01)
data02 = np.asarray(img02)
data03 = np.asarray(img03)

print(f'{data01.shape} {data02.shape} {data03.shape}')

#mix_image = (data01+data02+data03).clip(min = 0, max = 255)

A = np.array([[1, 0.5, 0.5],
              [0.5, 1, 0.5],
              [0.5, 0.5, 1]])

flat_data_01 = np.array(data01).reshape(40000, -1)
flat_data_02 = np.array(data02).reshape(40000, -1)
flat_data_03 = np.array(data03).reshape(40000, -1)

S = (np.c_[flat_data_01, flat_data_02, flat_data_03]).T

X = np.dot(A, S)

# display the recovered images
for i in range(X.shape[0]):
  print(X[i])
  rdata = X[i].reshape(200,200).clip(min=0, max=255).astype(int)
  print(rdata)
  img_mix = Image.fromarray(rdata)
  display(img_mix)
```
